```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# Scatter Plot: 
    # A plot of individual data points as dots, often used to display the relationship between two variables.

import matplotlib.pyplot as plt    
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.
This is particularly useful when you want to display multiple plots or charts together,
allowing you to compare different visualizations side by side or arrange them in a structured manner.
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1= np.array([0, 100, 200, 300, 400, 500])
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("subplot 1")

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.figure(figsize=(12,8))
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("subplot 2")


x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
plt.plot(x3,y3)
# plt.legend()
plt.title("subplot 3")


x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.figure(figsize=(12,8))
plt.subplot(2,2,4)
plt.plot(x1,y1)
plt.title("subplot 4")

plt.tight_layout()
plt.show()
```

```python
bar plot:
    A bar plot, also known as a bar chart or bar graph, is a type of data visualization that represents categorical data using rectangular bars. 
    Each bars length or height is proportional to the value it represents.
    Bar plots are commonly used to display and compare the frequency, count, or value of different categories.
    
Bar plots are particularly useful in the following scenarios:
    1. cetegorical comparison
    2. data distribution
    3. trends over time
    4. comparative analysis
    5. normal data representation
    
    
```
